```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df=pd.read_csv("/content/drive/My Drive/data/used_car.csv")
df
```

```
df=df.drop(["name"],axis=1)
```

```
df.isnull().sum()
```

```
print("seller       :",df.seller_type.unique())
print('year         :',df.year.unique())
print('transmission :',df.transmission.unique())
print("fuel type    :",df.fuel.unique())
print("owner        :",df.owner.unique())
```

```
final_df=pd.get_dummies(df,drop_first=True)
```

```
final_df
```

```
list(final_df.columns) 
```

```
final_df = final_df.rename(columns = {'owner_Fourth & Above Owner': 'fourth_owner', 'owner_Second Owner': 'second_owner', 'owner_Test Drive Car': 'first_owner', 'owner_Third Owner': 'third_owner'}, inplace = False)
```

```
X=final_df.drop(["selling_price"],axis=1)
y=final_df.iloc[:,1]
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```
from sklearn.ensemble import RandomForestRegressor
```

```
regressor=RandomForestRegressor(max_depth= 15,
 min_samples_leaf= 2,
 min_samples_split= 10,
 n_estimators=1100)
```

```
regressor.fit(X_train,y_train)
```

```
predictions=regressor.predict(X_test)
```

```
import pickle
# open a file, where you ant to store the data
file = open('used_car_regression_model.pkl', 'wb')

# dump information to that file
pickle.dump(regressor, file)
```
